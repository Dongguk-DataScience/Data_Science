```python
import numpy as np
import pandas as pd
import os
```

```python
def file_path(path):
  file_list = os.listdir(path)
  return file_list
```

```python
def into_one_csv(path, file_list, dict_name):
    df_total = pd.DataFrame()
    df_total["구역"] = 0

    for idx in range(len(file_list)):
        file_name = file_list[idx]
        df = pd.read_csv(path +'\\' + dict_name + '\\' + file_name)
        df_total=pd.concat([df_total,df])
    
    df_total["구역"] = dict_name

    return df_total
```

```python
path = '{type your path}'

dict_list = file_path(path)

print(dict_list)
```

```python
df = pd.DataFrame()
for dict_name in dict_list:
    file_list = file_path(path + '\\' + dict_name)
    df=pd.concat([df,into_one_csv(path,file_list,dict_name)])
```
